```python
## merging df_tmdb with df2 on 'id' column using outer join
# This will help in combining the data from both DataFrames based on the 'id' column 

merged_df_outer1 = pd.merge(df_tmdb, df2, on='id', how='outer')  
merged_df_outer1.isnull().sum()
```

```python
## checking the information about the merged DataFrame
# This will help in verifying the merge operation and understanding the structure of the DataFrame
merged_df_outer1.info()
```

```python
merged_df_outer1.drop(columns='original_language', inplace=True)
```

```python
# 1. Strip whitespace and standardize column names
merged_df_outer1.columns = merged_df_outer1.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# 2. Fill missing values

# Fill categorical/text columns with MODE
for col in [ 'movie_title', 'release_date']:
    if merged_df_outer1[col].isnull().any():
        mode_val = merged_df_outer1[col].mode()[0]
        merged_df_outer1[col].fillna(mode_val, inplace=True)
```

```python
# Fill numerical columns with MEAN
for col in ['popularity', 'vote_average']:
    if merged_df_outer1[col].isnull().any():
        mean_val = merged_df_outer1[col].mean()
        merged_df_outer1[col].fillna(mean_val, inplace=True)
```

```python
# Fill budget/gross with 0 or mean depending on your strategy
for col in ['production_budget', 'domestic_gross', 'worldwide_gross']:
    if merged_df_outer1[col].isnull().any():
        merged_df_outer1[col].fillna(merged_df_outer1[col].mean(), inplace=True)
```

```python
# 3. Drop rows where 'movie' is still missing if critical
merged_df_outer1.dropna(subset=['movie'], inplace=True)
```

```python

# 4. Filter out zero or negative budgets to avoid division errors
merged_df_outer1 = merged_df_outer1[merged_df_outer1['production_budget'] > 0]
```

```python

# 5. Compute ROI and ROI Percent
merged_df_outer1['roi'] = (merged_df_outer1['worldwide_gross'] - merged_df_outer1['production_budget']) / merged_df_outer1['production_budget']
merged_df_outer1['roi_percent'] = merged_df_outer1['roi'] * 100
# 6. Normalize ROI so that the total sums to 100%
total_roi = merged_df_outer1['roi'].sum()
```

```python
# 6. Normalize ROI % to sum to 100%
merged_df_outer1['normalized_roi_%'] = (merged_df_outer1['roi_percent'] / merged_df_outer1['roi_percent'].sum()) * 100
# 7. Remove any NaN or infinite ROI values (just in case)
merged_df_outer1 = merged_df_outer1[merged_df_outer1['roi'].notna() & merged_df_outer1['roi'].apply(np.isfinite)]   
```
